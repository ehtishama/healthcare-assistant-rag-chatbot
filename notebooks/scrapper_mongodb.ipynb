```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1536)
```

```python
from langchain_core.documents import Document
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    add_start_index=True,

)
```

```python
docs = []

# preprocessing
for link in links:
    page = requests.get(link)
    soup = BeautifulSoup(page.content, "html.parser")

    # all the relvant text exists in the section tags
    sections = soup.find_all("section")

    # scrap only those pages with at least 2 sections tags
    if len(sections) < 2:
        continue

    text = ""
    for section in sections:
        text = text + section.text.strip().replace("  ", "").replace("\n", "")

    doc = Document(page_content=text, metadata={"source": link, "title": soup.find("title").text.replace("\n", "")})
    docs.append(doc)


```
